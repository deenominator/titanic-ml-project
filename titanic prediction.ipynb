```python
women = train[(train["Sex"] == "female") & (train["Survived"] == 1)]
women2    = train[(train["Sex"] == "female") & (train["Survived"] == 0)]
```

```python
t_aw= len(women)
t_dw= len(women2)
sum= t_aw+t_dw
```

```python
rate_w= (t_aw/sum)*100
print(f"% of women survived= {rate_w}")
```
